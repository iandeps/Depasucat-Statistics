```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# load dataset
df = pd.read_csv(r'dataset/current-power.csv', 
                 delimiter=",")
df
```

```python
df.describe()
```

```python
df.cov()
```

```python
# scatter plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel( 'Current (A)')
plt.show
```

```python
# correlation coefficient of current and power
df.corr()
```

```python
# regression line 
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

# scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color = 'blue',
            alpha = 0.6, 
            label='Data Points')
# line plot
plt.plot(x, line,
         color = 'red',
         label=f'Regression Line (r = {r_value:.3f})')
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel( 'Current (A)')
plt.legend()
plt.show()
#hello
```
